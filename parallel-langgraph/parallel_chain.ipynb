```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class BatsmanState(TypedDict):

    runs: int
    balls: int
    fours: int
    sixes: int

    sr: float
    bpb: float
    boundary_percent: float
    summary: str
```

```python
def calculate_sr(state: BatsmanState):

    sr = (state['runs']/state['balls'])*100
    
    return {'sr': sr}
def calculate_bpb(state: BatsmanState):

    bpb = state['balls']/(state['fours'] + state['sixes'])

    return {'bpb': bpb}
def calculate_boundary_percent(state: BatsmanState):

    boundary_percent = (((state['fours'] * 4) + (state['sixes'] * 6))/state['runs'])*100

    return {'boundary_percent': boundary_percent}
```

```python
def summary(state: BatsmanState):

    summary = f"""
Strike Rate - {state['sr']} \n
Balls per boundary - {state['bpb']} \n
Boundary percent - {state['boundary_percent']}
"""
    
    return {'summary': summary}
```

```python
workflow = StateGraph(BatsmanState)

workflow.add_node('sr',calculate_sr)
workflow.add_node('bpb',calculate_bpb)
workflow.add_node('b_percent',calculate_boundary_percent)
workflow.add_node('summary',summary)

workflow.add_edge(START,'sr')
workflow.add_edge(START,'bpb')
workflow.add_edge(START,'b_percent')

workflow.add_edge('sr','summary')
workflow.add_edge('bpb','summary')
workflow.add_edge('b_percent','summary')

workflow.add_edge('summary',END)

graph = workflow.compile()
```

```python
intial_state = {
    'runs': 100,
    'balls': 50,
    'fours': 6,
    'sixes': 4
}

graph.invoke(intial_state)
```
