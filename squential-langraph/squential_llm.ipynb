```python
from langgraph.graph import StateGraph,START,END
from langchain_ollama import ChatOllama
import torch
from typing import TypedDict
```

```python
class LLMState(TypedDict):
    question:str
    answer:str
```

```python
def llm_qa(state:LLMState)->LLMState:
    device = torch.cuda.is_available()
    llm = ChatOllama(model="qwen3-vl:4b",temperature=0)
    print(device)
    prompt = f"Answer the following question: {state['question']}"
    answer = llm.invoke(prompt)
    state['answer'] = answer.content
    return state
```

```python
#define your graph
workflow = StateGraph(LLMState)

#add nodes
workflow.add_node('llm_qa',llm_qa)

#add edges
workflow.add_edge(START,'llm_qa')
workflow.add_edge('llm_qa',END)

#compile graph
graph = workflow.compile()
```

```python
initial = {'question':"What is capital of Delhi and can you tell the population of world"}
final_result = graph.invoke(initial)
print(final_result)
```
